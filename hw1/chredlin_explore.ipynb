Here, we do some exploratory analysis on the `chredlin` dataset.

```python
import numpy as np
import pandas as pd
from chredlin import load_data

np.set_printoptions(suppress=True)

chredlin = load_data()
chredlin
```

```python
import seaborn as sns

with sns.plotting_context('paper'):
    g = sns.PairGrid(chredlin.drop(columns=['income']),
                     diag_sharey=False, hue='side', height=2)
    g = g.map_diag(sns.distplot, kde=False)
    g = g.map_offdiag(sns.scatterplot)
    g.add_legend()
    g.fig.suptitle('chredlin')
    g.fig.subplots_adjust(top=0.95)
    g.savefig('p1_pair_plots.pdf')
```
