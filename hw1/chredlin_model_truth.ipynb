This is the model to replicate.

```r
library(faraway)
data(chredlin)
linear.model <- lm(involact ~ race + fire + theft + age + log(income),
                   data=chredlin)
summary(linear.model)
```

```r
chredlin
```

```r
predict(linear.model, chredlin[1,], interval = 'prediction')
```
